```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("DeepX_unlabeled.xlsx")
data
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
```

```python
print (data.isnull().sum())
```

```python
data.dropna(subset=['review_text'],inplace=True)
```

```python
def clean_text(text):
    text = str(text)
    text= text.lower()
    text = re.sub(r'http\s+|www\s+','',text)
    text = re.sub(r'\d+','',text)
    text = re.sub(r'[^\w\s]','',text)
    text = text.strip()
    return text
```

```python
data['cleaned_text'] = data['review_text'].apply(clean_text)
```

```python
def convert_date(text):
    text = str(text)
    if'hour'in text:
        return 0
    elif'day'in text:
        return 1
    elif'week'in text:
        return 7
    elif'month'in text:
        return 30
    else:
        return np.nan
```

```python
data['days_ago'] = data['date'].apply(convert_date)
```

```python
data['star_rating'] = data['star_rating'].astype(int)
```

```python
data['review_length'] = data['cleaned_text'].apply(len)
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
print("\nCleaned Data Preview:")
print(data.head())
```

```python
plt.figure()
sns.countplot(x='star_rating', data=data)
plt.title('Star Rating Distribution')
plt.xlabel('Star Rating')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
sns.histplot(data['review_length'], bins=30)
plt.title('Review Length Distribution')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()
```

```python
def get_sentiment(rating):
    if rating >= 4:
        return 'Positive'
    elif rating <= 2:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
data['sentiment'] = data['star_rating'].apply(get_sentiment)
```

```python
sentiment_counts = data[data['sentiment'] != 'Neutral']['sentiment'].value_counts()
```

```python
plt.figure()
```

```python
plt.pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Positive vs Negative Reviews')

plt.show()
```
